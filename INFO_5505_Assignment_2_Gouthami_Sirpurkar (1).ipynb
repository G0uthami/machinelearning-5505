```python
#Separate the Dataset into Input & Output NumPy Arrays
```

```python
# store dataframe values into a numpy array

array = input.values

# separate array into input and output by slicing
# for X(input) [:, 1:12] --> all the rows, columns from 1 - 11
# these are the independent variables or predictors

X = array[:,1:12]

# for Y(input) [:, 11] --> all the rows, column 11
# this is the value we are trying to predict

Y = array[:,11]
```

Seperate arrays were formed for the predictors or data used to predict the target variable, and the target which value to be predicted.

```python
#Spilt into Input/Output Array into Training/Testing Datasets
```

```python
# split the dataset --> training sub-dataset: 80%; test sub-dataset: 20%

test_size = 0.20

#selection of records to include in each data sub-dataset must be done randomly

seed = 7

#split the dataset (input and output) into training / test datasets

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size,
random_state=seed)
```

Test_size is set to 0.20 seed is set to 7 the data set is split into training and test data sets.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Build and Train the Model
```

```python
#build the model

model = LogisticRegression(random_state=seed, max_iter=1000)

# train the model using the training sub-dataset

model.fit(X_train, Y_train)

#print the classification report

predicted = model.predict(X_test)
report = classification_report(Y_test, predicted)
print("Classification Report: ", "\n", "\n",report)
```

The input dataset is split into two dataframes one for trianing purpose and the other for validation. The training data and testing data are split in 20 and 80 percent proportions respectively. Then the performance of the model is checked using the clasification report, the accuracy of the model and many other parameters such as precision, recall, and f1-score were also available to look at.

```python
#Score the Accuracy of the Model
```

```python
#score the accuracy level

result = model.score(X_test, Y_test)

#print out the results

print(("Accuracy: %.3f%%") % (result*100.0))
```

The accuracy of the model is checked and print function is used to print the result.

```python
#Evaluate the Model using the 10-fold Cross-Validation Technique.
```

```python
# Evaluate the algorithm and specify the number of times of repeated splitting, in this case 10 folds

n_splits=10

#Fix the random seed.  You must use the same seed value so that the same subsets can be obtained for each time the process is repeated

seed=7

kfold=KFold(n_splits, random_state=seed, shuffle=True)

# for logistic regression, we can use the accuracy level to evaluate the model 

scoring="accuracy"

#train the model and run K-fold cross validation to validate / evaluate the model

results=cross_val_score (model, X, Y, cv=kfold, scoring=scoring)

# print the evaluation results.  The result is the average of all the results obtained from the K-fold cross validation

print("Accuracy: %.3f (%.3f)"% (results.mean(), results.std()))
```

New model is trained using k-fold cross validation method and the accuracy for the new model is the mean of the all the results from k-fold which in this case is 10. And there is a slight increase in the accuracy for this model from the previous one.
